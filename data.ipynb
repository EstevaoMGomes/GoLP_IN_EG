# Introduction
To start the project first I am going to see to what extent PySR is capable by varying:
- seed
- error
- number of points


Let´s start with the function
$$ f(x_0,x_1)=2.1 e^{-(1.4x_0^2+4.3x_0x_1+x_1^2) }$$
for a given seed and with no error. 


```python
seed = np.array([1,2,3,4])
n = np.array([100,1000,10000,100000])
np.random.seed(seed[0])

def gaussian2D(A,a,b,c,x,y):
    return A*np.exp(-(a*x**2+b*x*y+c*y**2))

i=1
X = 6 * np.random.rand(n[i],2)-3
y = gaussian2D(2.1,1.4,4.3,1,X[:,0],X[:,1])
```

Starting our symbolic regression algorithm...

```python
model = PySRRegressor(
    niterations=20,
    populations=50,
    binary_operators=["+", "-", "*"],
    unary_operators=["exp"],
)
model.fit(X, y)
```

```python
model.sympy()
```

```python
for i in range(4):
    X = 6 * np.random.rand(n[i],2)-3
    y = gaussian2D(2.1,1.4,4.3,1,X[:,0],X[:,1])
    model.fit(X, y)
    model.sympy()
```
